```python
class Polinomio:
    "clase que define y realiza operaciones con polinomios\nSe definen colocando como argumento los coeficientes comenzando desde el de mayor orden"
    
    def __init__(self, coef=[]):
        self.coef=coef 
    
    def grado(self):
        "Devuelve el grado del polinomio"
        gr=len(self.coef)-1
        return gr
    
    def get_coeficientes(self):
        "Devuelve los coeficientes del polinomio en forma de lista, comenzando desde el mayor orden"
        return self.coef
    
    def set_coeficientes(self,lista):
        "Establece los coeficientes del polinomio en forma de lista.\nSe definen colocando como argumento los coeficientes comenzando desde el de mayor orden"
        self.coef=lista
        
    def suma_pol(self,pol):
        "suma dos polinomios, o un polinomio con una lista de coeficientes comenzando con el de mayor orden"
        resultado=Polinomio([])
        if type(pol) is list:
            pol=Polinomio(pol)
        if type(pol) is Polinomio:
            if len(self.coef)<len(pol.coef):
                for x in range(len(self.coef),len(pol.coef)):
                    self.coef.append(0)
            elif len(self.coef)>len(pol.coef):
                for x in range(len(pol.coef),len(self.coef)):
                    pol.coef.append(0)
            ncoef = [y +pol.coef[x] for x,y in enumerate(self.coef)]
            resultado.set_coeficientes(ncoef)
        else: print("Error de tipo, el argumento debe ser un polinomio o una lista numérica")
        return resultado
    
    def mul(self,cte):
        "Multiplica un polinomio con una constante y retorna otro polinomio"
        ncoef = [cte*x for x in self.coef]
        resultado=Polinomio(ncoef)
        return resultado
    
    def derivada(self,n=1):
        "Realiza la derivada n de un polinomio y retorna otro polinomio"
        ncoef=self.coef
        for i in range(n):
            ncoef=[(x*y) for x,y in enumerate(ncoef)]
            ncoef.pop(0)
        if len(ncoef)==0:
            ncoef=[0]
        resultado= Polinomio(ncoef)
        return resultado
    
    def integral(self,n=1,cte=0):
        "Realiza la antiderivada n de un polinomio, teniendo en cuenta la constante.\nRetorna otro polinomio"
        ncoef=self.coef
        lcte=[cte]
        for i in list(range(1,n+1)):
            ncoef=[(y/(x+1)) for x,y in enumerate(ncoef)]
            ncoef=lcte+ncoef
        resultado= Polinomio(ncoef)
        #print(lcte)
        return resultado
    
    def from_string(self,expr="",var="x"):
        "Genera un polinomio a partir de un string"
        expr=expr.lower()
        expr=expr.replace(" ","")
        expr=expr.replace("-","+-")
        term=expr.split("+")
        var=var.lower()
        #estandarizo la lísta de términos
        for i,j in enumerate(term):
            if var+"^" in j:
                if j.startswith(var):
                    j="1"+j
                elif j.startswith("-"+var):
                    j=j.replace("-"+var,"-1"+var)
            elif var in j:
                if j.startswith(var):
                    j="1"+j
                elif j.startswith("-"+var):
                    j=j.replace("-"+var,"-1"+var)
                j=j.replace(var,var+"^1")
            else:
                #independiente
                j=j+var+"^0"
            term[i]=j
            
        #sumo los términos en una lista
        lista=[0 for x in range(9)]
        for l in term:
            trm=l.split(var+"^")
            lista[int(trm[1])]+=float(trm[0])
            
        #saco los términos que sobran
        lista.reverse()
        for m in lista:
            if m==0.0:
                lista.pop(0)
            else:
                break
        lista.reverse()
        self.set_coeficientes(lista)
        
    def __str__(self):
        res=""
        coef=self.coef
        coef.reverse()
        l=len(coef)-1
       
        for i,j in enumerate(coef):
            if j!=0.0: # si es 0.0 no lo muestra
                if i< l: 
                    if coef[i+1]>=0: #Para que no ponga +-
                        if l-i>1:
                            res+=" "+str(j)+" x^"+str(l-i)+" +"
                        else:
                            res+=" "+str(j)+" x +"
                    else:
                        if l-i>1:
                            res+=" "+str(j)+" x^"+str(l-i)
                        else:
                            res+=" "+str(j)+" x"
                else:
                    res+=" " + str(j)#+"x^"+str(l-i)
        
        return res
        
    def __call__(self,x):
        res=0.0
        for n,t in enumerate(self.coef):
            res+=t*(x**n)
        return res
        
    def __add__(self,p):
        resultado=self.suma_pol(p)
        return resultado
    
    def __mul__(self,k):
        res=self.mul(k)
        return res
    
    def __rmul__(self,k):
        res=self.mul(k)
        return res
```
